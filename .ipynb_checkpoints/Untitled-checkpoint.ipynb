# Part 2: Feature extraction, visualization, and statistics

## Feature extraction

```python
# import libraries
import pandas as pd
import string
import collections
import nltk
```

```python
# read csv
df = pd.read_csv("thebiglist_clean.csv")
```

```python
df
```

```python
# get the Variety column
variety = df['Variety']
variety
```

```python
# join the Variety column into one long string
variety_text = " ".join(variety)
variety_text
```

```python
# make casing consistent
variety_text_lower = variety_text.lower()
variety_text_lower
```

```python
# remove punctuation
punc_list = list(string.punctuation)
punc_list
```

```python
# remove punctuations
variety_clean = "".join([char for char in variety_text_lower if char not in punc_list])
variety_clean
```

```python
variety_tokens = nltk.tokenize.word_tokenize(variety_clean)
variety_tokens
```

```python
# look into frequency of tokens
variety_counter = collections.Counter(variety_tokens)
variety_counter
```

```python
variety_counter.most_common()
```

```python
is_spicy = []

hot_flavor = ['spicy', 'curry', 'hot', 'tom', 'chilli', 'chili', 'kari', 'pedas', 'laksa']

for variety in df['Variety'].str.lower():
    if any(flavor in variety for flavor in hot_flavor):
        is_spicy.append(1)
    else:
        is_spicy.append(0)
        
df['isSpicy'] = is_spicy
df
```

```python
has_chicken = []

for variety in df['Variety'].str.lower():
    if('chicken' in variety) or ('ayam' in variety):
        has_chicken.append(1)
    else:
        has_chicken.append(0)

df['hasChicken'] = has_chicken
df
```

```python
df.sample(10)
```

```python
has_beef = []

for variety in df['Variety'].str.lower():
    if('beef' in variety) or ('cow' in variety):
        has_beef.append(1)
    else:
        has_beef.append(0)

df['hasBeef'] = has_beef
df
```

```python
df.sample(10)
```

```python
seafood_flavor = ['seafood', 'fish', 'shrimp', 'prawn', 'crab']

has_seafood = []

for variety in df['Variety'].str.lower():
    if any(flavor in variety for flavor in seafood_flavor):
        has_seafood.append(1)
    else:
        has_seafood.append(0)
        
df['hasSeafood'] = has_seafood
df
```

## Questions to ask:
1. Which brand has the highest stars?
2. Which country has the highest stars?
3. Which noodle has higher stars - spicy or non-spicy

```python
# Question 1: which brand has the highest stars?
plt.figure(figsize = (16,9))
sns.boxplot(df['Brand'], df['Stars'], order = brand_count.index[:20], showfliers = False)
plt.xticks(rotation = 45)
```

### By median, MyKuali ranks the best 

```python
# explore MyKuali
df[df['Brand'] == 'MyKuali']
```

```python
# Question 2: Which country has the highest stars?
plt.figure(figsize = (16,9))
sns.boxplot(df['Country'], df['Stars'], order = country_count.index[:20], showfliers = False)
plt.xticks(rotation = 45)
```

### By median, Malaysia seems to perform the best

```python
# Question 3: Which noodles has higher stars - spicy or non-spicy
sns.boxplot(df['isSpicy'], df['Stars'], showfliers = False)
```

```python
# filter 2 dataframes
spicy_df = df[df['isSpicy'] == 1]
nonspicy_df = df[df['isSpicy'] == 0]
```

```python
print(spicy_df['Stars'].mean())
print(nonspicy_df['Stars'].mean())
```

```python
print(spicy_df['Stars'].median())
print(nonspicy_df['Stars'].median())
```

```python
# perform statistical test to see if the difference is significant
sns.distplot(spicy_df['Stars'], color = 'red')
sns.distplot(nonspicy_df['Stars'], color = 'black')
```

```python
# import mann-whitney u test
from scipy.stats import mannwhitneyu
```

```python
mannwhitneyu(spicy_df['Stars'], nonspicy_df['Stars'])
```

### The small p-value of < 0.05 shows that there is sufficient evidence to reject the null hypothesis and that there is a statistical significance in the difference between stars of spicy and stars of non-spicy

### Hence spicy noodles has higher stars than nonspicy noodles

```python
# explore stars for chicken
sns.boxplot(df['hasChicken'], df['Stars'], showfliers = False)
```

```python
mannwhitneyu(df[df['hasChicken']==0]['Stars'], df[df['hasChicken'] == 1]['Stars'])
```

```python
# explore stars for beef
sns.boxplot(df['hasBeef'], df['Stars'], showfliers = False)
```

```python
# explore stars for seafood
sns.boxplot(df['hasSeafood'], df['Stars'], showfliers = False)
```
